```python
import pandas as pd
%matplotlib inline
```

```python
air = pd.read_csv("Daqi.csv", parse_dates=True)
air.head(10)
```

```python
air['site'] = air['site'].str.replace('\d+', '').str.strip()
```

```python
air.plot()
```

```python
air.pivot(columns="site", values="index_value").head()
```

```python
air.groupby("site").mean().reset_index()
```

This dataset shows the air quality index as shown in each county in Pittsburgh. Not only does it show air quality index, but also health advisory and a description of who the air quality would affect. This dataset connects to our metric of environmentally friendly because the lower the air quality index, the better air quality and environmentally friendly for habitants. The county with the least air quality is the most environmentally friendly.
